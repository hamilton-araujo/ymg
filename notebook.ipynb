```python
import numpy as np

df_temp = pd.merge(
    abundancias_ymg, 
    valores_medidos[['id', 'feh', 'err_feh_', 'discoespesso']], 
    on='id', 
    how='left'
)

# 4. CÃ¡lculo de [Mg/Fe] e sua incerteza
# [Mg/Fe] = [MgI] - feh
df_temp['[Mg/Fe]'] = df_temp['[MgI]'] - df_temp['feh']
df_temp['e_[Mg/Fe]'] = np.sqrt(df_temp['e_[MgI]']**2 + df_temp['err_feh_']**2)

# 5. CÃ¡lculo de [Y/Mg] e sua incerteza
# [Y/Mg] = [YII] - [MgI] (Note: abundÃ¢ncia relativa Ã© geralmente secundÃ¡rio - primÃ¡rio)
# Se vocÃª deseja especificamente [MgI] - [YII] como solicitado:
df_temp['[Y/Mg]'] = df_temp['[YII]'] - df_temp['[MgI]']
df_temp['e_[Y/Mg]'] = np.sqrt(df_temp['err_YII']**2 + df_temp['e_[MgI]']**2)

# 6. Atualizar o dataframe original abundancias_ymg com as novas colunas
abundancias_ymg = df_temp.copy()

# Exibir as primeiras linhas para conferÃªncia
print(abundancias_ymg[['id', '[Mg/Fe]', 'e_[Mg/Fe]', '[Y/Mg]', 'e_[Y/Mg]']].head())

# Salvar o novo arquivo CSV
abundancias_ymg.to_csv('abundancias_ymg_atualizado.csv', index=False)
```

```python
# 2. DefiniÃ§Ã£o da funÃ§Ã£o da reta por partes
def mg_fe_model(feh):
    if feh <= -0.6:
        return 0.19
    elif -0.6 < feh <= 0:
        return -0.217 * feh + 0.06
    else: # feh > 0
        return 0.06

# Criar pontos para a linha do modelo
x_line = np.linspace(abundancias_ymg['feh'].min() - 0.05, abundancias_ymg['feh'].max() + 0.05, 500)
y_line = [mg_fe_model(x) for x in x_line]

# --- SEPARAÃ‡ÃƒO DOS DADOS ---
mask_espesso = abundancias_ymg['discoespesso'] == True
espesso = abundancias_ymg[mask_espesso]
fino = abundancias_ymg[~mask_espesso]

# 3. Criar o grÃ¡fico
plt.figure(figsize=(10, 8))

# Plotar as barras de erro (de todos os pontos de uma vez, para simplificar)
plt.errorbar(
    abundancias_ymg['feh'], 
    abundancias_ymg['[Mg/Fe]'], 
    xerr=abundancias_ymg['err_feh_'], 
    yerr=abundancias_ymg['e_[Mg/Fe]'],
    fmt='none', 
    ecolor=(0.7, 0.7, 0.7, 0.4), 
    elinewidth=1, 
    capsize=2,
    zorder=1
)

# Plotar os pontos do Disco Fino (cÃ­rculos 'o')
scatter_fino = plt.scatter(
    fino['feh'], 
    fino['[Mg/Fe]'], 
    s=80, 
    c=fino['age_mean'],
    cmap='viridis',
    marker='o',
    edgecolor='black', 
    linewidth=0.5,
    alpha=1, 
    label='Disco Fino',
    zorder=2,
    vmin=abundancias_ymg['age_mean'].min(), # Garante que a escala de cores seja igual
    vmax=abundancias_ymg['age_mean'].max()
)

# Plotar os pontos do Disco Espesso (estrelas '*')
scatter_espesso = plt.scatter(
    espesso['feh'], 
    espesso['[Mg/Fe]'], 
    s=150, # Estrela precisa ser um pouco maior para compensar o formato
    c=espesso['age_mean'],
    cmap='viridis',
    marker='*',
    edgecolor='black', 
    linewidth=0.5,
    alpha=1, 
    label='Disco Espesso',
    zorder=3,
    vmin=abundancias_ymg['age_mean'].min(),
    vmax=abundancias_ymg['age_mean'].max()
)

# Plotar a reta do modelo
plt.plot(
    x_line, 
    y_line, 
    color='red', 
    linewidth=2.5, 
    label='DivisÃ£o QuÃ­mica (Modelo)', 
    zorder=4
)

# Linha de referÃªncia no zero
plt.axhline(0, color='black', linestyle='--', linewidth=1, alpha=0.5, zorder=0)

# Colorbar (usamos o scatter_fino como referÃªncia, pois vmin/vmax sÃ£o iguais)
cbar = plt.colorbar(scatter_fino)
cbar.set_label('Idade Isocronal (Gyr)', fontsize=12)

# ConfiguraÃ§Ãµes de eixos e tÃ­tulos
plt.xlabel('[Fe/H] (dex)')
plt.ylabel('[Mg/Fe] (dex)')
plt.title('DistinÃ§Ã£o entre Disco Fino e Disco Espesso com Idade')
plt.legend(scatterpoints=1) # Garante que apareÃ§a apenas um sÃ­mbolo na legenda
plt.grid(True, linestyle=':', alpha=0.6)

# Salvar
plt.savefig('mg_fe_vs_feh_final.png')
```

```python
import plotly.express as px
import plotly.graph_objects as go

def mg_fe_model(feh):
    if feh <= -0.6:
        return 0.19
    elif -0.6 < feh <= 0:
        return -0.217 * feh + 0.06
    else: # feh > 0
        return 0.06

x_line = np.linspace(abundancias_ymg['feh'].min() - 0.1, abundancias_ymg['feh'].max() + 0.1, 500)
y_line = [mg_fe_model(x) for x in x_line]

fig = px.scatter(
    abundancias_ymg, 
    x='feh', 
    y='[Mg/Fe]',
    error_x='err_feh_',        # Barras de erro no eixo X
    error_y='e_[Mg/Fe]',      # Barras de erro no eixo Y
    hover_name='id',    
    color= 'age_mean',      # Mostra o ID da estrela ao passar o mouse
    labels={
        'feh': '[Fe/H] (dex)',
        '[Mg/Fe]': '[Mg/Fe] (dex)',
        'age_mean': 'Idade MÃ©dia Isocronal (Gyr)'
    },
    title='EvoluÃ§Ã£o QuÃ­mica: [Mg/Fe] vs [Fe/H]',
    template='plotly_white',
    color_continuous_scale='Viridis',
    # --- PARÃ‚METROS DE TAMANHO ---
    width=1000,   # Largura em pixels
    height=800    # Altura em pixels
)

fig.update_traces(
    marker=dict(
        size=12,          # Aumenta o tamanho dos pontos
        opacity=0.85,     # Leve transparÃªncia nos pontos
        line=dict(width=1, color='DarkSlateGrey') # Borda para destacar
    ),
    # Define cor RGBA (Red, Green, Blue, Alpha) para as barras de erro
    error_x=dict(color='rgba(128, 128, 128, 0.3)', thickness=1), 
    error_y=dict(color='rgba(128, 128, 128, 0.3)', thickness=1),
    selector=dict(mode='markers')
)

fig.add_trace(go.Scatter(
    x=x_line, 
    y=y_line, 
    mode='lines', 
    name='Modelo por Partes', 
    line=dict(color='red', width=2.5)
))

# Adicionar linha de referÃªncia em zero
fig.add_hline(y=0, line_dash="dash", line_color="black", opacity=0.5)

# Salvar o grÃ¡fico interativo como arquivo HTML
fig.write_html('mg_fe_vs_feh_plotly.html')

# Mostrar o grÃ¡fico (em ambientes compatÃ­veis)
fig.show()
```

```python
a, e_a = -27.701, 0.461
b, e_b = 3.454, 0.329
c, e_c = 4.576, 1.013

# 5. CÃ¡lculo da Idade (Age_ymg)
abundancias_ymg['age_ymg'] = a * abundancias_ymg['[Y/Mg]'] + b * abundancias_ymg['feh'] + c

# 6. PropagaÃ§Ã£o do Erro (err_age_ymg)
term_x = (a * abundancias_ymg['e_[Y/Mg]'])**2    # em relaÃ§Ã£o a [Y/Mg]
term_y = (b * abundancias_ymg['err_feh_'])**2    # em relaÃ§Ã£o a [Fe/H]
term_a = (abundancias_ymg['[Y/Mg]'] * e_a)**2    # em relaÃ§Ã£o ao coeficiente a
term_b = (abundancias_ymg['feh'] * e_b)**2       # em relaÃ§Ã£o ao coeficiente b
term_c = (1 * e_c)**2               # em relaÃ§Ã£o Ã  constante c
abundancias_ymg['err_age_ymg'] = np.sqrt(term_x + term_y + term_a + term_b + term_c)

# 7. Salvar o resultado
abundancias_ymg.to_csv('abundancias_ymg_com_idade.csv', index=False)

# Visualizar as primeiras linhas
print(abundancias_ymg[['id', 'age_ymg', 'err_age_ymg']].head())
```

```python
import pandas as pd
import numpy as np

# 1. DefiniÃ§Ã£o da funÃ§Ã£o com propagaÃ§Ã£o de erro
def calcular_idade_quimica_2(ymg, e_ymg):
    # Coeficientes e suas incertezas
    a, e_a = -27.6243, 0.3053
    b, e_b = 2.3950, 0.8139
    
    # CÃ¡lculo do valor central: y = a*x + b
    age = (a * ymg) + b
    
    # PropagaÃ§Ã£o de erro: sigma_y = sqrt( (x*sigma_a)^2 + (a*sigma_x)^2 + (sigma_b)^2 )
    e_age = np.sqrt((ymg * e_a)**2 + (a * e_ymg)**2 + (e_b)**2)
    
    return age, e_age

# 2. Aplicar apenas para as estrelas na faixa: -0.71 <= [Fe/H] < -0.17
mask_feh = (abundancias_ymg['feh'] >= -0.71) & (abundancias_ymg['feh'] < -0.17)

# Inicializar as colunas novas
abundancias_ymg['age_ymg2'] = np.nan
abundancias_ymg['err_age_ymg2'] = np.nan

# 3. Executar o cÃ¡lculo
resultados = abundancias_ymg[mask_feh].apply(
    lambda x: calcular_idade_quimica_2(x['[Y/Mg]'], x['e_[Y/Mg]']), axis=1
)

# Separar os resultados em duas colunas
if not resultados.empty:
    abundancias_ymg.loc[mask_feh, 'age_ymg2'] = [r[0] for r in resultados]
    abundancias_ymg.loc[mask_feh, 'err_age_ymg2'] = [r[1] for r in resultados]

# 4. FormataÃ§Ã£o para exibiÃ§Ã£o (Valor (Erro))
def formatar(val, err):
    if pd.isna(val): return "-"
    # Usando a mesma lÃ³gica de arredondamento anterior para 2 algarismos significativos no erro
    from math import log10, floor
    exp = floor(log10(abs(err)))
    prec = exp - 1
    dec = max(0, -prec)
    return f"{round(val, dec):.{dec}f} ({round(err, dec):.{dec}f})".replace('.', ',')

abundancias_ymg['age_ymg2_fmt'] = abundancias_ymg.apply(
    lambda x: formatar(x['age_ymg2'], x['err_age_ymg2']), axis=1
)

# Exibir resultados para as estrelas que entraram na faixa
print(abundancias_ymg[mask_feh][['id', 'feh', '[Y/Mg]', 'age_ymg2_fmt']].to_string(index=False))
```
